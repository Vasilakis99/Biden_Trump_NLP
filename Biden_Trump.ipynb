```python
import pandas as pd
import re
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from torch.nn.functional import softmax
from tqdm import tqdm
import emoji
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
!pip install langdetect
```

```python
# Text preprocessing 

def preprocess_text(text):
    text = text.lower()
    text = emoji.replace_emoji(text, replace='')  # remove emojis
    text = re.sub(r"http\S+|www\S+|https\S+", '', text)  # remove URLs
    text = re.sub(r"[^\w\s.,!?']", '', text)  # keep common punctuation
    text = re.sub(r"\s+", ' ', text).strip()  # remove excessive whitespace
    return text
```

```python
# Preprocess comments

df_b["clean_comment"] = df_b["Comment Text Original"].astype(str).apply(preprocess_text)
df_t["clean_comment"] = df_t["Comment Text Original"].astype(str).apply(preprocess_text)
```

```python
from langdetect import detect, DetectorFactory
from langdetect.lang_detect_exception import LangDetectException

DetectorFactory.seed = 0  # ensures consistent results

def detect_language(text):
    try:
        return detect(text)
    except LangDetectException:
        return "error"

# Apply language detection
df_b["language"] = df_b["clean_comment"].apply(detect_language)

# Count and compare
biden_lang_counts = df_b["language"].value_counts(normalize=True) * 100

print("üîµ Biden Dataset Language Distribution üîµ (%):")
print(biden_lang_counts.head(10))
```

```python
from langdetect import detect, DetectorFactory
from langdetect.lang_detect_exception import LangDetectException

DetectorFactory.seed = 0  # ensures consistent results

def detect_language(text):
    try:
        return detect(text)
    except LangDetectException:
        return "error"

# Apply language detection
df_t["language"] = df_t["clean_comment"].apply(detect_language)

# Count and compare
trump_lang_counts = df_t["language"].value_counts(normalize=True) * 100

print("üî¥ Trump Dataset Language Distribution üî¥ (%):")
print(trump_lang_counts.head(10))
```

```python
#Import model & tokenise 

model_name = "j-hartmann/emotion-english-roberta-large"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
model.eval()
```

```python
# Emotion labels for this model
emotion_labels = ['anger', 'disgust', 'fear', 'joy', 'neutral', 'sadness', 'surprise']

def predict_emotion(texts, batch_size=16):
    predictions = []
    for i in tqdm(range(0, len(texts), batch_size)):
        batch = texts[i:i+batch_size]
        tokens = tokenizer(batch, padding=True, truncation=True, return_tensors="pt", max_length=512)
        
        with torch.no_grad():
            outputs = model(**tokens)
            probs = softmax(outputs.logits, dim=1)
            labels = torch.argmax(probs, dim=1)
            predictions.extend([emotion_labels[label] for label in labels])
    return predictions
```

```python
# Predict emotions
df_b["emotion"] = predict_emotion(df_b["clean_comment"].tolist())
```

```python
df_t["emotion"] = predict_emotion(df_t["clean_comment"].tolist())
```

```python
# Combine for comparative visualization

biden_emotions = df_b["emotion"].value_counts(normalize=True).rename("Biden")
trump_emotions = df_t["emotion"].value_counts(normalize=True).rename("Trump")

emotion_df = pd.concat([biden_emotions, trump_emotions], axis=1).fillna(0)
emotion_df = emotion_df.reindex(['joy', 'sadness', 'anger', 'fear', 'disgust', 'surprise', 'neutral'])

# Bar plot: side-by-side comparison
emotion_df.plot(kind="bar", figsize=(10,6))
plt.title("Emotion Distribution in YouTube Comments")
plt.ylabel("Proportion of Comments")
plt.xlabel("Emotion Category")
plt.xticks(rotation=45)
plt.legend(title="Candidate")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Pie charts for each candidate
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

df_b["emotion"].value_counts().plot.pie(autopct='%1.1f%%', ax=axes[0], title="Biden Comments")
df_t["emotion"].value_counts().plot.pie(autopct='%1.1f%%', ax=axes[1], title="Trump Comments")

plt.tight_layout()
plt.show()
```
